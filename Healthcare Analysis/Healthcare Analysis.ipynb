```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
sdf=pd.read_csv('/content/drive/MyDrive/dataset/student_scores_IDS_FA2.csv')
```

```python
sdf.head()
```

```python
sdf.info()
```

```python
x_features=sdf[['Hours']]
x=sm.add_constant(x_features)
y=sdf[['Scores']]
```

```python
s_model=sm.OLS(y,x).fit()
s_model.summary()
```

```python
y_pred=s_model.predict([[1,5.1]])
y_pred
```
